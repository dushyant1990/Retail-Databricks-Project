```python
from pyspark.sql import functions as F

df_orders_raw = spark.read.json("/Volumes/workspace/default/raw_landing/orders/")

df_orders_bronze = (
    df_orders_raw
    .withColumn("_ingested_at", F.current_timestamp())
    .withColumn("_source_file", F.element_at(F.split(F.col("_metadata.file_path"), "/"), -1))
)
#df_orders_bronze.display()
df_orders_bronze.write.format("delta").mode("append").saveAsTable("workspace.bronze.orders")
```
